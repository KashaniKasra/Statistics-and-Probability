```python
import numpy as np, random
import matplotlib.pyplot as mpl

def set_seed(seed):
    np.random.seed(seed)
    random.seed(seed)
set_seed(810109203)

#          Q3
def Regression(x, y):
    n = float(len(x))
    xy_sum = np.sum(x * y)
    a = ((n * xy_sum) - (np.sum(x) * np.sum(y))) / ((n * np.sum(x * x)) - (np.sum(x) ** 2))
    b = (np.sum(y) - (a * np.sum(x))) / n
     
    
    sum_x = np.sum(x)
    sum_y = np.sum(y)
    numerator = n * np.sum(x * y) - (sum_x * sum_y)
    coefficient = (numerator ** 2) / ((n * np.sum(x * x) - (sum_x ** 2)) * (n * np.sum(y * y) - (sum_y ** 2)))
    
    print("R2 =", coefficient)
    
    mpl.plot(x, y)
    mpl.axline((0, b), slope = a)
    mpl.show()


# original
x1 = np.array([-2.3, -1.1, 0.5, 3.2, 4.0, 6.7, 10.3, 11.5])
y1 = np.array([-9.6, -4.9, -4.1, 2.7, 5.9, 10.8, 18.9, 20.5])
Regression(x1, y1)

# outlietr
x2 = np.array([-2.3, -1.1, 0.5, 3.2, 4.0, 6.7, 10.3, 11.5, 5.8])
y2 = np.array([-9.6, -4.9, -4.1, 2.7, 5.9, 10.8, 18.9, 20.5, 31.3])
Regression(x2, y2)

# high leverage
x3 = np.array([-2.3, -1.1, 0.5, 3.2, 4.0, 6.7, 10.3, 11.5, 20.4])
y3 = np.array([-9.6, -4.9, -4.1, 2.7, 5.9, 10.8, 18.9, 20.5, 14.1])
Regression(x3, y3)

# both
x4 = np.array([-2.3, -1.1, 0.5, 3.2, 4.0, 6.7, 10.3, 11.5, 20.4])
y4 = np.array([-9.6, -4.9, -4.1, 2.7, 5.9, 10.8, 18.9, 20.5, 31.3])
Regression(x4, y4)
```
